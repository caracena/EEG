```python
import numpy
import random
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def getTimeSeries(n,i,f):
    cont = []
    for x in range(n):
        cont.append(random.randint(i, f))
    return stats.zscore(cont)
```

```python
def getBackward(seq,k):
    return numpy.diff(seq,n=k)
```
